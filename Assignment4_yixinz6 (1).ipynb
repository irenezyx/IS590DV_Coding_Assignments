### What I've Implemented
* Draw a Heat Map of Total Square Footage with Agency Name and Congress Dist.
* When selected a grid or multiple grids of the heat map, draw the corresponding line plot of Total Square Footage VS. Year Acquired.
* Change the line plot when unselected the grids or select different/more grids in heat map.
* It's also ok to change Congress Dist with a slider and Agency Name with a drop box (without selecting the heat map) to purely see the change of line plot.
* The figures are in the bottom of this notebook. Selection information will also be output.

```python
import pandas as pd
import bqplot
import numpy as np
import ipywidgets
import traitlets
import matplotlib.pyplot as plt
from IPython.display import clear_output
# can use type() to check type
```

```python
buildings = pd.read_csv("/home/shared/sp18-is590dv/data/IL_Building_Inventory.csv",
                       na_values = {'Year Acquired':0, 'Year Constructed':0, 'Square Footage':0,
                                   'Agency Name': 'Missed Value', 'Congress Dict': 'Missed Value'})
#buildings
```

```python
data = buildings.groupby(['Agency Name', 'Congress Dist'])['Square Footage'].sum()
data
```

```python
data = data.unstack().fillna(0)
data
```

```python
data.index.values # index object
```

```python
data.keys() # return Index object
```

```python
def new_plot(agency, congress): 
    if type(agency) is not list:
        agency = [agency]
    if type(congress) is not list:
        congress = [congress]
    df2 = buildings.loc[( buildings['Congress Dist'].isin(congress) 
               & buildings['Agency Name'].isin(agency))]
    y = df2.groupby('Year Acquired')['Square Footage'].sum()
    x = list(df2.groupby('Year Acquired').groups.keys())
    clear_output(True)
    if len(x) == 0 :
        print('Empty plot as there\'s no square footage for this congress dist and agency!')
    plt.gcf().clear() # this function might need to be changed to plt.clear() with different environment
    plt.xlabel('Year Acquired')
    plt.ylabel('Total Square Footage')
    plt.title('Total Square Footage VS. Year Acquired of Selected Agency Name and Congress Dist')
    plt.plot(x, y, '-bo')
    plt.show()
```

```python
col_sc = bqplot.ColorScale(scheme = 'Greens') # scheme red
x_sc = bqplot.OrdinalScale()
y_sc = bqplot.OrdinalScale()
y_ax = bqplot.Axis(scale = y_sc, label = data.index.name, 
                   orientation = "vertical")
x_ax = bqplot.Axis(scale = x_sc, label = data.keys().name)
y_ax.tick_values = data.index.values
c_ax= bqplot.ColorAxis(scale=col_sc, orientation='vertical', side = 'right') # set color bars
heat_map = bqplot.GridHeatMap(
                            color = data, 
                            row = data.index.values,
                            column = data.keys().values,
                            scales = {'color':col_sc,
                                     'row': y_sc,
                                     'column': x_sc},
                            interactions = {'click': 'select'},
                            selected_style={'opacity':1.2}, 
                            unselected_style={'opacity':0.2}
                             ) 
# now click on a block can select one with ctrl to unselect; with shift to select area 
fig = bqplot.Figure(marks=[heat_map], 
                    axes=[x_ax, y_ax, c_ax],
                    fig_margin = {'top': 100, 'bottom': 60, 'left': 300, 'right' :100},
                    title = 'Heat Map of Total Square Footage with Agency Name and Congress Dist'
                   )
fig.layout.width = '900px'
fig.layout.height = '650px'


a_label = ipywidgets.Label()
d_label = ipywidgets.Label()
f_label = ipywidgets.Label()
out = ipywidgets.Output()

def selection_changed(event):
    if event['new'] is None:  # for unselect operation
        with out:
            clear_output(True)
            a_label.value = 'Agency Name: No selected item.'
            d_label.value = 'Congress Dist: No selected item.'
            f_label.value = 'Total Square Footage: No selected item.'
            print('Unselected!')
        return
    
    agency_name = []
    congress_dist = []
    footage = 0
    a_label.value = 'Agency Name: '
    d_label.value = 'Congress Dist: '
    for i, j in event['new']: # note that we might select mutiple blocks
        agency_name.append(data.index.values[i])
        congress_dist.append(data.keys().values[j])
        footage += data[j][i]
        a_label.value += (str(data.index.values[i]) + ' ')
        d_label.value += (str(data.keys().values[j]) + ' ')
    f_label.value = 'Total Square Footage: ' + str(footage)
    
    with out:
        clear_output(True)
        if len(agency_name) > 1 or len(congress_dist) > 1:
            new_plot(agency_name, congress_dist)
        else:
            slider = ipywidgets.IntSlider(min=0, max=18, step=1, value=congress_dist[0])
            agency_list = ipywidgets.Dropdown(
                options = data.index.values,
                value = agency_name[0]
            )
            plot = ipywidgets.interactive(new_plot, agency=agency_list,
                                      congress=slider)
            display(plot)
            
        
heat_map.observe(selection_changed, 'selected')

display(ipywidgets.VBox([fig, a_label, d_label, f_label, out]))
```
