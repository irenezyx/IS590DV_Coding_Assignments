```python
import pandas as pd
import bqplot
import ipywidgets
import json
```

```python
films = json.load(open("shared/sp18-is590dv/data/star_wars/films.json"))
```

```python
starships_input = json.load(open("shared/sp18-is590dv/data/star_wars/starships.json"))
```

```python
char_input = json.load(open("shared/sp18-is590dv/data/star_wars/people.json"))
characters = {}
for character in char_input:
    characters[character['pk']] = character['fields']
```

```python
for film in films:
    ch_id = film['fields']['characters'][0]
    film_name = film["fields"]["title"]
    print(film_name, ch_id, characters[ch_id]['name'])
```

```python
for i in characters:
    print(characters[i]['name'])
print(characters[1])
```

```python
char_to_media = {}
media_to_char = {}
for film in films:
    for j in range(len(film['fields']['characters'])):
        ch_id = film['fields']['characters'][j]
        film_name = film["fields"]["title"]
        cha_name = characters[ch_id]['name']
        
        if cha_name not in char_to_media:
            char_to_media[cha_name] = [film_name]
        else:
            if film_name not in char_to_media[cha_name]:
                char_to_media[cha_name].append(film_name)
                
        if film_name not in media_to_char:
            media_to_char[film_name] = [cha_name]
        else:
            if cha_name not in media_to_char[film_name]:
                media_to_char[film_name].append(cha_name)
media_to_char
```

```python
node_data = []
node_id = {}
i = 0
for char in char_to_media:
    node_data.append({'label': char, 'media': char_to_media[char]})
    node_id[char] = i
    i+= 1
node_id
```

```python
link_data = []
linked_map = []
for i in range(len(char_to_media)):
    tmp = []
    for j in range(len(char_to_media)):
        tmp.append(0)
    linked_map.append(tmp)
    linked_map[i][i] = 1
for film in media_to_char:
    chars = media_to_char[film]
    for i in range(len(chars)):
        for j in range(i+1, len(chars)):
            if linked_map[i][j]:
                continue
            link_data += [{'source': node_id[chars[i]], 'target': node_id[chars[j]]},
                         {'source': node_id[chars[j]], 'target': node_id[chars[i]]}]
            linked_map[i][j] = linked_map[j][i] = 1
link_data
```

```python
tooltip = bqplot.Tooltip(fields=["label", "media"])

graph = bqplot.Graph(node_data=node_data, link_data=link_data, tooltip = tooltip, highlight_links = True,
                    colors = ['#9EB6B8']*len(char_to_media))

fig = bqplot.Figure(marks = [graph])
fig.layout.width = '900px'
fig.layout.height = '650px'

from IPython.display import clear_output

instr = ipywidgets.Textarea('Selected node will be highlighted to orange and related nodes will be highlighted to yellow.')
selected_label = ipywidgets.Textarea('Selected Characters: Nobody')
related_label = ipywidgets.Textarea('Characters who have appeared in the same movie with selected character: No related nodes as nothing selected')

def selection_changed(event):
    if event['new'] is None:
        return
    clear_output(True)
    ID = event['new'][0]
    tmp = ['#9EB6B8']*len(node_data)
    linked_nodes = []
    tmp[ID] = 'orange'
    for i in range(len(link_data)):
        if link_data[i]['source'] == ID:
            tmp[link_data[i]['target']] = 'yellow'
            linked_nodes.append(node_data[link_data[i]['target']]['label'])
    graph.colors = tmp
    fig = bqplot.Figure(marks = [graph])
    fig.layout.width = '900px'
    fig.layout.height = '650px'
    global selected_label, related_label
    selected_label.value = 'Selected character: {0}'.format(node_data[ID]['label'])
    related_label.value = 'Characters who have appeared in the same movie with selected character: '
    flag = False
    for char in linked_nodes:
        if flag:
            related_label.value += ', '
        related_label.value += char
        flag = True
    if not flag:
        related_label.value += 'No other characters share movies with it!'
    display(ipywidgets.VBox([ipywidgets.HBox([instr, selected_label, related_label]), fig]))
graph.observe(selection_changed, 'selected')

display(ipywidgets.VBox([ipywidgets.HBox([instr, selected_label, related_label]), fig]))
```
